## Ciclo While
#1. Contar del 1 al 5

##Descripción:
##El ciclo while repite instrucciones mientras una condición sea verdadera.

##Situación:
##Mostrar números consecutivos en pantalla.

```python
# Programa realizado por Andres Jaquez NL = 27

# Se solicita hasta qué número quiere contar el usuario
limite = int(input("Ingresa el número límite: "))

# Variable contador
contador = 1

# El ciclo continuará mientras el contador sea menor o igual al límite
while contador <= limite:

    # Se muestra el valor actual del contador
    print("Número:", contador)

    # El contador aumenta en 1
    contador = contador + 1
print("Programa realizado por Andres Jaquez NL = 27")
```

#2. Tabla de multiplicar con while

##Descripción:
##El ciclo while puede repetirse varias veces para mostrar operaciones matemáticas.

##Situación:
##Mostrar la tabla de multiplicar de un número.

```python
# Programa realizado por Andres Jaquez NL = 27

# Se solicita un número
numero = int(input("Ingresa un número: "))

# Variable contador
contador = 1

# El ciclo se repetirá hasta el número 10
while contador <= 10:

    # Se realiza la multiplicación
    resultado = numero * contador

    # Se muestra el resultado
    print(numero, "x", contador, "=", resultado)

    # El contador aumenta en 1
    contador = contador + 1
print("Programa realizado por Andres Jaquez NL = 27")
```

#3. Suma de números

##Descripción:
##El ciclo while permite acumular valores repetidamente.

##Situación:
##Sumar varios números ingresados por el usuario.

```python
# Programa realizado por Andres Jaquez NL = 27

# Variable para guardar la suma
suma = 0

# Variable contador
contador = 1

# El ciclo se repetirá 5 veces
while contador <= 5:

    # Se solicita un número
    numero = int(input("Ingresa un número: "))

    # El número se agrega a la suma
    suma = suma + numero

    # El contador aumenta en 1
    contador = contador + 1

# Se muestra la suma total
print("La suma total es:", suma)
print("Programa realizado por Andres Jaquez NL = 27")
```

#4. Contar letras de una palabra

##Descripción:
##El ciclo while puede recorrer caracteres de una palabra usando posiciones.

##Situación:
##Mostrar letra por letra una palabra.

```python
# Programa realizado por Andres Jaquez NL = 27

# Se solicita una palabra
palabra = input("Ingresa una palabra: ")

# Variable contador
contador = 0

# El ciclo continuará mientras el contador sea menor al tamaño de la palabra
while contador < len(palabra):

    # Se muestra cada letra de la palabra
    print(palabra[contador])

    # El contador aumenta en 1
    contador = contador + 1
print("Programa realizado por Andres Jaquez NL = 27")
```

#5. Ciclo while

##Descripción:
##El ciclo while repite instrucciones mientras una condición sea verdadera.
##Situación: Contar números del 1 al 5.

```python
# Programa realizado por Andres Jaquez NL = 27

# Se solicita un número inicial
numero = int(input("Ingresa el número inicial para comenzar la cuenta regresiva: "))

# El ciclo continuará mientras el número sea mayor o igual a 1
while numero >= 1:

    # Se muestra el número actual
    print("Cuenta regresiva:", numero)

    # El número disminuye en 1
    numero = numero - 1

# Mensaje final
print("Fin de la cuenta regresiva")
print("Programa realizado por Andres Jaquez NL = 27")
```
